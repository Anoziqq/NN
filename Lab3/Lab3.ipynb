```python
import numpy as np 
from sklearn import datasets 
import matplotlib.pyplot as plt 
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import confusion_matrix
```

## Генерація набору даних

```python
X, y = datasets.make_classification(n_samples=2000,n_informative=1, n_features=2, n_classes=2, n_redundant=0, n_clusters_per_class=1,class_sep = 2) 
fig, ax = plt.subplots(figsize=(9, 6)) 
plt.xlabel("X0", fontsize=20) 
plt.ylabel("X1", fontsize=20) 
plt.scatter(X[:,0], X[:,1], s=50, c=y)
```

## Моделювання наборів даних та візуалізація резулататів

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
min1, max1 = X[:, 0].min()-1, X[:, 0].max()+1
min2, max2 = X[:, 1].min()-1, X[:, 1].max()+1
x1_array = np.arange(min1, max1, 0.01)
x2_array = np.arange(min2, max2, 0.01)
xx, yy = np.meshgrid(x1_array, x2_array)
r1, r2 = xx.flatten(), yy.flatten()
r1, r2 = r1.reshape((len(r1), 1)), r2.reshape((len(r2), 1))
grid = np.hstack((r1,r2))

model = LinearDiscriminantAnalysis()
model.fit(X, y)
y_predict = model.predict(grid)
zz = y_predict.reshape(xx.shape)
plt.contourf(xx,yy,zz,cmap='Pastel2')
plt.scatter(X[:,0], X[:,1], s=50, c=y)
plt.show()
```

## Значення метрик якості

```python
TP, TN, FP, FN,  = confusion_matrix(y,model.predict(X)).ravel()
print('True Positive(TP)  = ', TP)
print('True Negative(TN)  = ', TN)
print('False Positive(FP) = ', FP)
print('False Negative(FN) = ', FN)
```
